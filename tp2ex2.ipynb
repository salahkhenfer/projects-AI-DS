###  Add  tauxendettement column

```python
df_im['tauxendettement'] = (df_im['remboursement'] * 100) / df_im['revenu']
df_im
```

###  Number of People Exceeding Debt Ratio 35%

```python
langht_tauxendettement =   len(df_im[(df_im['tauxendettement'] > 35)])
langht_tauxendettement
```

###  Number of people in Algiers exceeding the debt ratio of 35%

```python
num_people_exceeding_ratio_algiers = len(
    df_im[(df_im["tauxendettement"] > 35) & (df_im["ville"] == "ALGER")]
)
print(
    f"Number of people in Algiers exceeding debt ratio: {num_people_exceeding_ratio_algiers}"
)
```

### Adding the 'risk' column, marking clients as high-risk (1) if debt ratio > 35%

```python
#df_im["risk"] = df_im[df_im['tauxendettement'] > 35]
df_im["risk"] = df_im['tauxendettement'] > 35

df_im
```

### Adding the 'total' column,

```python
df_im['count_total'] = df_im['remboursement'] * df_im['duree']

df_im
```

### Adding the 'benefices' column,

```python
df_im['benefices'] = (df_im['count_total'] * df_im['taux'] / 100) / 24
df_im
```

 ### count the number of auto loans and calculate the average cost for those loans

```python
# Filter for auto loans
auto_loans = df_im[df_im['type'] == "automobile"]
auto_loans
print(f"Average total cost of auto loans: {len(auto_loans)}")

# Calculate total cost and average
average_auto_loan_cost = auto_loans["count_total"].mean()
print(f"Average total cost of auto loans: {average_auto_loan_cost}")
```

### esumming  the  duree telling in week  so we should using  this way  

```python

ouargla_loans = df_im[df_im["ville"] == "OUARGLA"].copy()

# Calculate total cost for each loan
ouargla_loans["count_total_month"] = ouargla_loans["remboursement"] * 4


# Calculate the profit for each loan
ouargla_loans["benefices"] = (
    ouargla_loans["count_total_month"] * ouargla_loans["taux"] / 100
) / 24

total_monthly_profit = ouargla_loans["benefices"].sum()

print(f"Total monthly profit for Ouargla: {total_monthly_profit}")
ouargla_loans
```
